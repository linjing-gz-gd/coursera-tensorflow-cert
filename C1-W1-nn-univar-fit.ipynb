```python
import numpy as np
import matplotlib.pyplot as plt
import utilities as utl
import tensorflow as tf
from tensorflow import keras

tf.__version__
```

```python
rng = np.random.default_rng()
```

```python
def hw_function(x):
    return 2 * x - 1
```

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
xs = np.arange(-1, 5, 1.)
ys = hw_function(xs) + rng.uniform(-0.5, 0.5, xs.shape)
```

```python
%%time

model.fit(xs, ys, epochs=500, verbose=1)
```

```python
model.predict([10.])
```

```python
xs_pred = np.arange(-10, 10, 0.5)
ys_pred = model.predict(xs_pred)
ys_true = hw_function(xs_pred)
```

```python
fig, ax = plt.subplots()

ax.plot(xs_pred, ys_true, label='true')
ax.plot(xs, ys, 'o', label='data')
ax.plot(xs_pred, ys_pred, label='pred')

utl.set_ax(ax, 'x', 'y', 'one-layer NN fit', legend=True)
```

```python
model2 = keras.Sequential([
    keras.layers.Dense(units=10, input_shape=(1,), activation=tf.nn.relu),
    keras.layers.Dense(units=1)])
model2.compile(loss='mean_squared_error', optimizer='sgd')
```

```python
xs = np.linspace(0, 2*np.pi, 50)
ys = np.sin(xs)
```

```python
model2.fit(xs, ys, epochs=500, verbose=0)
```

```python
xs_pred = np.linspace(-2*np.pi, 4*np.pi, 151)
ys_pred = model2.predict(xs_pred)
ys_true = np.sin(xs_pred)
```

```python
fig, ax = plt.subplots()

ax.plot(xs_pred, ys_true, label='true')
ax.plot(xs, ys, 'o', label='data')
ax.plot(xs_pred, ys_pred, label='pred')

utl.set_ax(ax, 'x', 'y', 'one-layer NN fit', legend=True)
```
